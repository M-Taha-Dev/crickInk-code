```python
import math
import numpy as np
import pandas as pd

#first we need to identify the regions in which batsman plays the most
def finding_optimal_field(data):
    region1 = ['DeepFineLeg','LongLeg','DeepSquareLeg','ShortFineLeg','BackwardSquareLeg','SquareLeg']
    region2 = ['DeepMidWicket','CowCorner','LongOn','MidOn','MidWicket','Straight']
    region3 = ['LongOff', 'DeepExtraCover', 'DeepCover','MidOff','ExtraCover','Cover']
    region4 = ['DeepPoint', 'DeepBackwardPoint', 'ThirdMan', 'Point', 'BackwardPoint','ShortThirdMan','FlySlip']


    r1 = ['LongStop','DeepFineLeg','LongLeg','DeepSquareLeg']
    r2 = ['DeepMidWicket','CowCorner','LongOn','Straight']
    r3 = ['LongOff', 'DeepExtraCover', 'DeepCover']
    r4 = ['DeepPoint', 'DeepBackwardPoint', 'ThirdMan']

    fields = r1 + r2 + r3 + r4
    field_positions = {
        "BehindSquare": "8.0",
        "Cover": "16.0",
        "CowCorner": "24.0",
        "DeepCover": "28.0",
        "DeepFineLeg": "28.0",
        "DeepMidWicket": "172.0",
        "DeepPoint": "52.0",
        "DeepSquareLeg": "72.0",
        "ExraCover": "4.0",
        "LongOff": "48.0",
        "LongOn": "24.0",
        "MidWicket": "24.0",
        "Point": "40.0",
        "Square": "12.0",
        "SquareLeg": "8.0",
        "StraightHit": "32.0",
        "ThirdMan": "56.0",
        "shortCover": "4.0"
    }
    temp_arr = []

    for key in field_positions:
        if key in fields:
            temp_arr.append(key)




    #        "Delivery_Length": "GoodLength Middle",
    def find_strength(data):
        region1_strength = 0.0
        region2_strength = 0.0
        region3_strength = 0.0
        region4_strength = 0.0
        for key,value in data.items():
            data[key] = float(value)
            if key in region1:
                region1_strength += data[key]
            elif key in region2:
                region2_strength += data[key]
            elif key in region3:
                region3_strength += data[key]
            elif key in region4:
                region4_strength += data[key]
        return [region1_strength,region2_strength,region3_strength,region4_strength]

    def cal_field_number(data):
        array = find_strength(data)
        divider = sum(array)
        for i in range(len(array)):
            array[i] = round(array[i]/divider * 9)
        while 0 in array:
            index = array.index(max(array))
            array[index] -= 1
            index2 = array.index(0)
            array[index2] += 1
        while sum(array) != 9:
            index = array.index(min(array))
            array[index] += 1
        return array

    region1 = ['LongStop', 'DeepFineLeg', 'LongLeg', 'DeepSquareLeg', 'ShortFineLeg', 'BackwardSquareLeg', 'SquareLeg']
    region2 = ['DeepMidWicket', 'CowCorner', 'LongOn', 'MidOn', 'MidWicket', 'Straight']
    region3 = ['LongOff', 'DeepExtraCover', 'DeepCover', 'MidOff', 'ExtraCover', 'Cover']
    region4 = ['DeepPoint', 'DeepBackwardPoint', 'ThirdMan', 'Point', 'BackwardPoint', 'ShortThirdMan', 'FlySlip']

    # Combine all regions into one list
    all_field_positions = region1 + region2 + region3 + region4

    # Initialize the initData dictionary with all positions set to False
    init_data = {position: False for position in all_field_positions}

    # Now, init_data contains all field positions with their values initialized to False

    # Initialize an empty dictionary to represent the adjacency list
    adjacency_list = {}

    # Function to add an edge to the graph
    def add_edge(u, v):
        if u in adjacency_list:
            adjacency_list[u].append(v)
        else:
            adjacency_list[u] = [v]

    # Adding edges to the adjacency list
    add_edge('DeepSquareLeg', 'DeepMidWicket')
    add_edge('DeepSquareLeg', 'LongLeg')
    add_edge('DeepSquareLeg', 'SquareLeg')
    add_edge('LongLeg', 'DeepSquareLeg')
    add_edge('LongLeg', 'DeepFineLeg')
    add_edge('LongLeg', 'BackwardSquareLeg')
    add_edge('DeepFineLeg', 'LongStop')
    add_edge('DeepFineLeg', 'LongLeg')
    add_edge('DeepFineLeg', 'ShortFineLeg')
    add_edge('LongStop', 'DeepFineLeg')
    add_edge('LongStop', 'ThirdMan')
    add_edge('ThirdMan', 'LongStop')
    add_edge('ThirdMan', 'DeepBackwardPoint')
    add_edge('ThirdMan', 'ShortThirdMan')
    add_edge('DeepBackwardPoint', 'BackwardPoint')
    add_edge('DeepBackwardPoint', 'ThirdMan')
    add_edge('DeepBackwardPoint', 'DeepPoint')
    add_edge('DeepPoint', 'Point')
    add_edge('DeepPoint', 'DeepBackwardPoint')
    add_edge('DeepPoint', 'DeepCover')
    add_edge('DeepCover', 'DeepPoint')
    add_edge('DeepCover', 'DeepExtraCover')
    add_edge('DeepCover', 'Cover')
    add_edge('DeepExtraCover', 'LongOff')
    add_edge('DeepExtraCover', 'DeepCover')
    add_edge('DeepExtraCover', 'ExtraCover')
    add_edge('LongOff', 'Straight')
    add_edge('LongOff', 'DeepExtraCover')
    add_edge('LongOff', 'MidOff')
    add_edge('Straight', 'LongOff')
    add_edge('Straight', 'LongOn')
    add_edge('LongOn', 'Straight')
    add_edge('LongOn', 'CowCorner')
    add_edge('LongOn', 'MidOn')
    add_edge('CowCorner', 'DeepMidWicket')
    add_edge('CowCorner', 'LongOn')
    add_edge('DeepMidWicket', 'CowCorner')
    add_edge('DeepMidWicket', 'DeepSquareLeg')
    add_edge('DeepMidWicket', 'MidWicket')
    add_edge('MidOn', 'LongOn')
    add_edge('MidOff', 'LongOff')
    add_edge('ExtraCover', 'DeepExtraCover')
    add_edge('Cover', 'DeepCover')
    add_edge('Point', 'DeepPoint')
    add_edge('BackwardPoint', 'DeepBackwardPoint')
    add_edge('ShortThirdMan', 'ThirdMan')
    add_edge('ShortFineLeg', 'DeepFineLeg')

    add_edge('BackwardSquareLeg', 'LongLeg')
    add_edge('SquareLeg', 'DeepSquareLeg')
    add_edge('MidWicket', 'DeepMidWicket')

    add_edge('ShortFineLeg', 'BackwardSquareLeg')
    add_edge('BackwardSquareLeg', 'ShortFineLeg')
    add_edge('BackwardSquareLeg', 'SquareLeg')
    add_edge('SquareLeg', 'BackwardSquareLeg')

    add_edge('MidOff', 'ExtraCover')
    add_edge('ExtraCover', 'MidOff')
    add_edge('ExtraCover', 'Cover')
    add_edge('Cover', 'ExtraCover')
    add_edge('SquareLeg', 'BackwardSquareLeg')

    # Displaying the adjacency list

    def check_adjacent(position1, position2):
        if position1 in adjacency_list and position2 in adjacency_list[position1]:
            return True
        return False


    check_adjacent('DeepMidWicket','LongOn')

    def are_adjacent(position1, position2):
        return position1 in adjacency_list and position2 in adjacency_list[position1]

    def find_non_adjacent_combinations(positions,count, current_combination=[]):
        if len(current_combination) == count:
            yield current_combination
        elif not positions:
            return
        else:
            for i, position in enumerate(positions):
                if all(not are_adjacent(position, p) for p in current_combination):
                    new_combination = current_combination + [position]
                    new_positions = positions[:i] + positions[i + 1:]
                    yield from find_non_adjacent_combinations(new_positions,count, new_combination)

    def field_combinations(region1, count):
        # List of field positions in region1

        # Create an empty list to store valid combinations
        valid_combinations = []

        # Find and store all valid combinations
        for combination in find_non_adjacent_combinations(region1, count):
            valid_combinations.append(combination)
        input_arr = np.array(valid_combinations)
        arr = np.sort(input_arr)
        unique_tuples = set(map(tuple, arr)) 

        # Convert tuples back to arrays
        unique_arrays = [np.array(t) for t in unique_tuples]
        arr = []
        for value in unique_arrays:
            arr.append(value)
        return arr


    field_combinations(region2,2)


    #finding field strengths
    numbers = cal_field_number(data)

    array1 = []
    array2 = []
    array3 = []
    array4 = []
    for i in range(len(numbers)):
        if i == 0:
            array1.append(field_combinations(region1, numbers[i]))
        elif i == 1:
            array2.append(field_combinations(region2, numbers[i]))
        elif i == 2:
            array3.append(field_combinations(region3, numbers[i]))                     
        elif i == 3:
            array4.append(field_combinations(region4, numbers[i]))


    array1= np.array(array1)
    array2= np.array(array2)
    array3= np.array(array3)
    array4= np.array(array4)


    c1 = []
    c2 = []
    c = []
    for val1 in array1[0]:
        for val2 in array2[0]:
            temp = np.concatenate((val1,val2))
            c1.append(temp)

    for val1 in array3[0]:
        for val2 in array4[0]:
            temp = np.concatenate((val1,val2))
            c2.append(temp)


    #result = np.concatenate((array1, array2), axis=0)
    for val1 in c1:
        for val2 in c2:
            temp = np.concatenate((val1,val2))
            c.append(temp)

    def find_valid(position1, position2):
        if position1 in adjacency_list and position2 in adjacency_list[position1]:

            return True
        return False


    def remove_invalid(array):
        count = 0
        r1 = ['LongStop','DeepFineLeg','LongLeg','DeepSquareLeg']
        r2 = ['DeepMidWicket','CowCorner','LongOn','Straight']
        r3 = ['LongOff', 'DeepExtraCover', 'DeepCover']
        r4 = ['DeepPoint', 'DeepBackwardPoint', 'ThirdMan']

        r = r1 + r2 + r3 + r4
        final = []
        for item in array:
            intersection_result = list(set(item).intersection(r))
            if len(intersection_result) < 5:
                final.append(item)
        return final

    # Filter out arrays with adjacent elements

    list_of_combinations = remove_invalid(c)
    """for item in list_of_combinations:
        print(item)"""
    df = pd.read_csv('./prd1.csv')
    def evaluate_field_combinations(data, field_combinations):
        results = []
        for combination in field_combinations:
            total_runs = 0
            for index, row in data.iterrows():
                shot_position = row['Shot_Field_Position']
                if shot_position not in combination:
                    total_runs += row['Runs']
            results.append((combination, total_runs))
        return results

    evaluation_results = evaluate_field_combinations(df, list_of_combinations)

    # Sort the results by total runs conceded
    sorted_results = sorted(evaluation_results, key=lambda x: x[1])

    # Print the sorted results
    for i, (combination, total_runs) in enumerate(sorted_results, 1):
        print(f"{i}. Field Combination: {combination} | Total Runs Conceded: {total_runs}")
        break
```
